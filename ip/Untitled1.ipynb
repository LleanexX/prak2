### 1.3.1 Задание
Создать 8x8 матрицу и заполнить её в шахматном порядке нулями и
единицами.

```python
import numpy as np
arr = np.zeros(64).reshape(8, 8)

for i in range(8):
    for j in range(8):
        if i % 2 == 0:
            if j % 2 == 0:
                arr[i][j] = 1
        else:
            if j % 2 != 0:
                arr[i][j] = 1
print(arr)
```

### 1.3.2 Задание
Создать 5x5 матрицу со значениями в строках от 0 до 4. Для создания
необходимо использовать функцию arrange.

```python
arr = np.empty((0, 0))
for _ in range(5):
    arr = np.append(arr, np.arange(0, 5))
arr = arr.reshape(5, 5)
print(arr)
```

### Задание 3.2.2
Загрузить фрейм данных по ссылке:
https://raw.githubusercontent.com/akmand/datasets/master/iris.csv.
Необходимо выполнить нормализацию первого числового признака
(sepal_length_cm) с использованием минимаксного преобразования, а
второго (sepal_width_cm) с задействованием z-масштабирования.

```python
import numpy as np
import pandas as pd
from sklearn import preprocessing
df = pd.read_csv('raw.csv')

length = list(df['sepal_length_cm'])
length = list(map(lambda x: [x], length))
length = np.array(length)

minmax = preprocessing.MinMaxScaler(feature_range = (0, 1))
minmax = minmax.fit_transform(length)
minmax
```

```python
width = list(df['sepal_width_cm'])
width = list(map(lambda x: [x], width))
width = np.array(width)

scaler = preprocessing.StandardScaler()
standard = scaler.fit_transform(width)
standard
```
